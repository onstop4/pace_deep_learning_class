```python
import flash
import torch
from flash.core.data.utils import download_data
```

```python
from flash.text import TextClassificationData, TextClassifier
```

```python
# replace xy12345n with your student ID
EXPERIMENT_NAME = "mlflow_autotracking_xy12345n"
mlflow.set_experiment(EXPERIMENT_NAME)
experiment = mlflow.get_experiment_by_name(EXPERIMENT_NAME)
print("experiment_id:", experiment.experiment_id)
REGISTERED_MODEL_NAME = 'classifier_model'
```

```python
mlflow.pytorch.autolog()
```

```python
print('### download IMDb data to local folder')
download_data("https://pl-flash-data.s3.amazonaws.com/imdb.zip", "./data/")
datamodule = TextClassificationData.from_csv(
    "review",
    "sentiment",
    train_file="data/imdb/train.csv",
    val_file="data/imdb/valid.csv",
    test_file="data/imdb/test.csv",
    batch_size=4,
)
```

```python
print('### define a text classifier ')
classifier_model = TextClassifier(backbone="prajjwal1/bert-tiny", num_classes=datamodule.num_classes)
```

```python
print('### define the trainer')
trainer = flash.Trainer(max_epochs=1, gpus=torch.cuda.device_count())
```

```python
print('### fine tune the pretrained model to get a new model for sentiment classification')
#replace xy12345n wigh your student ID
with mlflow.start_run(experiment_id=experiment.experiment_id, run_name="xy12345n_train") as dl_model_tracking_run:
  trainer.finetune(classifier_model, datamodule=datamodule, strategy="freeze")
```

```python
run_id = dl_model_tracking_run.info.run_id
print("run_id: {}; lifecycle_stage: {}".format(run_id,
                                               mlflow.get_run(run_id).info.lifecycle_stage))
```

```python
# register the fine-tuned model
logged_model = f'runs:/{run_id}/model'
mlflow.register_model(logged_model, REGISTERED_MODEL_NAME)
```

```python
data = TextClassificationData.from_lists(
    predict_data=[
        "Best movie I have seen.",
        "What a movie!",
    ],
    batch_size=4,
)
#replace xy12345n wigh your student ID
with mlflow.start_run(experiment_id=experiment.experiment_id, run_name="xy12345ntest"):
  predictions = trainer.predict(classifier_model, datamodule=data, output="labels")
print(predictions)
```
